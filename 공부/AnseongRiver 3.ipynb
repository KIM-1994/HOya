### **2. 시각화 준비**

+ 필요한 모듈을 라이브러리 한다.
+ 시각화 하기전에 필요한것들을 준비한다.
+ 한글이 깨지지않게 폰트와 마이너스값 에러 방지를 위해 코드를 입력해준다.

```python
%matplotlib inline 
# 주피터노트북에 바로 시각화 할 수 있게끔 나오게 하는것

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 사전준비
sns.set_context('talk') # 프리젠테이션이나 보고서와 같은 다양한 매체에 활용할 수 있도록 스타일링
sns.set_style('white') # background color, grid, spine, tick을 정의하여 그림의 전반적인 모양을 스타일링

# 한글 사용 준비
plt.rcParams['font.family'] = 'Malgun Gothic' # rcParams 전체 전역으로 한글 설정
plt.rcParams['axes.unicode_minus'] = False # 전체 전역으로 마이너스 값 에러 방지 (False는 확인선 제거)
```

+ 먼저, 측정하고싶은 BOD,COD 농도의 시각화를 만든다.

```python
fig, ax = plt.subplots(figsize=(20, 9))

# BOD농도
ax.plot(anseong3["년"], anseong3["BOD"], label = "BOD")

# COD농도
ax.plot(anseong3["년"], anseong3["COD"], label = "COD")

# 범례 추가
ax.legend()
```

+ 구분하기 쉽도록 점선과 실선으로 구분해준다.

```python
fig, ax = plt.subplots(figsize=(20, 9))

# BOD농도
ax.plot(anseong3["년"], anseong3["BOD"],'o-', label = "BOD")

# COD농도
ax.plot(anseong3["년"], anseong3["COD"],'o:', label = "COD")

# 범례 추가
ax.legend()
```

+ y축 숫자표현하기

```python
fig, ax = plt.subplots(figsize=(20, 9))

# BOD농도
ax.plot(anseong3["년"], anseong3["BOD"],'o-', label = "BOD")

# COD농도
ax.plot(anseong3["년"], anseong3["COD"],'o:', label = "COD")

# yticklabel 표현 수정
yticks = ax.get_yticks()
print(yticks)
ax.set_yticks(yticks)
yticklabels = ax.get_yticklabels()
print(yticklabels)
ax.set_yticklabels([y for y in yticklabels]) # yticklabels 사용

ax.legend()
```

+ 비교할 다른 데이터를 막대로 만듭니다.
+ 두 가지 그래프가 동시에 나오지 않는다는 것을 확인 할 수 있습니다.

```python
fig, ax = plt.subplots(figsize=(20, 9))

# pandas bar plot 추가
ax = anseong3[["수소이온농도","용존산소"]].plot.bar(ax=ax)

# BOD농도
ax.plot(anseong3["년"], anseong3["BOD"],'o-', label = "BOD")

# COD농도
ax.plot(anseong3["년"], anseong3["COD"],'o:', label = "COD")

# 범례 추가
ax.legend()
```

+ 잘 나오는지 두 개를 동시에 확인해 봅니다.

```python
fig, ax = plt.subplots(ncols=2,figsize=(20, 9), sharey = True)

# pandas bar plot 추가
anseong3[["수소이온농도","용존산소"]].plot.bar(ax=ax[0])

# BOD농도
ax[1].plot(anseong3["년"], anseong3["BOD"],'o-', label = "BOD")

# COD농도
ax[1].plot(anseong3["년"], anseong3["COD"],'o:', label = "COD")
```

+ y축의 값들이 동일하다는것을 확인할 수 있습니다.

```python
yticks0 = ax[0].get_xticks() 
yticks1 = ax[1].get_xticks()
print(f"yticks of ax[0] = {yticks0}")
print(f"yticks of ax[1] = {yticks1}")
```

+ y축을 통일 시킨 후 x의 값들을 동일하게 해줍니다.

```python
fig, ax = plt.subplots(ncols=2,figsize=(20, 9), sharey = True)

# pandas bar plot 추가
anseong3[["년","수소이온농도","용존산소"]].plot.bar(x= "년", y = ["수소이온농도","용존산소"],ax=ax[0], zorder = 0)

# BOD농도
ax[1].plot(anseong3["년"], anseong3["BOD"],'o-', label = "BOD")

# COD농도
ax[1].plot(anseong3["년"], anseong3["COD"],'o:', label = "COD")
```

+ 두 개의 그래프를 동시에 확인해 볼 수 있습니다.

```python
fig, ax = plt.subplots(figsize=(20, 9), sharey=True)

# pandas bar plot 추가
anseong3[["년","수소이온농도","용존산소"]].plot.bar(x= "년", y = ["수소이온농도","용존산소"],ax=ax)

# BOD농도
ax.plot(anseong3["BOD"],'o-', label = "BOD", color = 'r')

# COD농도
ax.plot(anseong3["COD"],'o:', label = "COD", color = 'g')



ax.legend()
```
